```python
from modules.data_types import Transmission
from modules.utils import get_colormap
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.patches import Patch as MPatch
from matplotlib import rcParams
```

```python
# Load the data.
# This is the same saved LDA plot data file from the "cell_types" Mesmerize project
t = Transmission.from_hdf5('./data/lda_on_hier_all_promoters.ptrn')
data = np.vstack(t.df._LDA_TRANSFORM.values)
```

# Fig 2f

```python
cmap = get_colormap(t.df.FCLUSTER_LABELS, 'tab10', alpha=0.6)
colors = list(map(cmap.get, t.df.FCLUSTER_LABELS))

fig = plt.figure(figsize=(2.6, 2.3), dpi=300)
ax = plt.subplot(111)

ax.scatter(x=data[:, 0], y=data[:, 1], c=colors, s=3, linewidths=0)

ax.tick_params(axis='both', which='major', labelsize=3, width=.1)
plt.setp(ax.spines.values(), linewidth=0.1)

ax.set_xlabel('LD1', fontdict={'fontsize': 4})
ax.set_ylabel('LD2', fontdict={'fontsize': 4})

ax.set_ylim([-4, 5])

# handles = [MPatch(color=cmap[k], label=k) for k in cmap.keys()]
# ax.legend(handles=handles, ncol=int(np.sqrt(len(handles))), loc='lower right')
fig.tight_layout()
```
